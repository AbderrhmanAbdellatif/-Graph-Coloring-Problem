```python
from unittest import result

class GraphColor:
    def __init__(self, V):
        self.adj = [[] for i in range(V)]
        self.V = V

    def addEdge(self, v, w):
        self.adj[v].append(w)
        self.adj[w].append(v)


    def greedyColoring(self):

        result = [-1] * self.V
        result[0] = 0
        available = [False] * self.V
        mxColor = 1
        for u in range(1, self.V):
            for i in self.adj[u]:
                if (result[i] != -1):
                    available[result[i]] = True

            cr = 0
            while cr < self.V:
                if (available[cr] == False):
                    break

                cr += 1
                mxColor = max(mxColor, cr+1)
            result[u] = cr
            for i in self.adj[u]:
                if (result[i] != -1):
                    available[result[i]] = False
        return (mxColor, result)
    
    def printColors(self):
        res = self.greedyColoring()
        print(res[0])
        print(*res[1])

# Driver Code
if __name__ == '__main__':
    f = open("/kaggle/input/graphcoloringproblem/sample1.txt")
    data = f.read()
    first_line = data.split('\n', 1)[0] # read first line
    P, V , E = first_line.strip().split()
    V = int(V)
    E = int(E)
    g = GraphColor(V)
    for line in lines:
        e, u, v = line.rstrip().strip().split()
        u = int(u) -1
        v = int(v) -1
        g.addEdge(u, v)
    g.printColors()
```
